```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KDTree
```

```python
data = pd.read_csv("C:\\research\\Analysis\\dataset-result-good-design-only-labeled.csv", engine='python', dtype={"src_image_text": "string", "background_image_text": "string"})
sample_poor_design = pd.read_csv("C:\\research\\Analysis\\sample-poor-design.csv", engine='python', dtype={"src_image_text": "string", "background_image_text": "string"})
combined = pd.concat([data, sample_poor_design], ignore_index=True, sort=False)
```

```python
features = [x for x in data.columns if (x in ["category", "permission"]) or ("wordvector" in x)]
features_need_encoding = [x for x in data.columns if x in ["category", "permission"]]
features_no_need_encoding = [x for x in data.columns if x not in ["id", "category","elemID","permission","layout_width","layout_height","textColor","text","textSize","src_image_text","background_image_text"]]
features_ui_element = [x for x in data.columns if x in ["id", "category","elemID","permission","layout_width","layout_height","textColor","text","textSize","src_image_text","background_image_text"]]
```

```python
le_dict = {}
for feat in features_need_encoding:
    lbl_enc = preprocessing.LabelEncoder()
    combined[feat] = lbl_enc.fit(combined[feat].fillna("-1").astype(str).values)
    le_dict[feat] = dict(zip(lbl_enc.classes_, lbl_enc.transform(lbl_enc.classes_)))
    
    data[feat] = lbl_enc.transform(data[feat].fillna("-1").astype(str).values)
    sample_poor_design[feat] = lbl_enc.transform(sample_poor_design[feat].fillna("-1").astype(str).values)

for feat in features_no_need_encoding:
    data[feat] = data[feat].fillna(0.0)
```

```python
X = data[features]
```

```python
tree = KDTree(X)
```
